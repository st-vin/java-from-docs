# Java Object-Oriented Programming Practice Questions

## EASY LEVEL

### Exercise 1: Basic Class Declaration and Member Variables
**Question:** Create a class called `Student` with the following member variables:
- `name` (String)
- `age` (int)
- `grade` (char)

Make all variables public for now.

### Exercise 2: Simple Constructor
**Question:** Add a constructor to the `Student` class that takes three parameters (name, age, grade) and initializes all member variables.

### Exercise 3: Creating and Using Objects
**Question:** Create a `main` method that creates two `Student` objects and prints their information.

## MEDIUM LEVEL

### Exercise 4: Methods with Return Values
**Question:** Add the following methods to your `Student` class:
- `getFullInfo()` - returns a String with all student information
- `isAdult()` - returns true if age >= 18
- `getGradePoint()` - converts letter grade to numeric (A=4, B=3, C=2, D=1, F=0)

### Exercise 5: Access Control
**Question:** Modify the `Student` class to make all member variables private. Add appropriate getter and setter methods for each variable. Include validation in the setters (age must be positive, grade must be A, B, C, D, or F).

### Exercise 6: Multiple Constructors and Method Overloading
**Question:** Create a `BankAccount` class with:
- Private variables: `accountNumber` (String), `balance` (double), `ownerName` (String)
- Three constructors:
  1. Takes all three parameters
  2. Takes only accountNumber and ownerName (balance starts at 0)
  3. Takes only ownerName (generates random account number, balance starts at 0)
- Overloaded `deposit` methods:
  1. `deposit(double amount)` - adds to balance
  2. `deposit(double amount, String description)` - adds to balance and prints description

## HARD LEVEL

### Exercise 7: Static Members and Complex Object Interaction
**Question:** Create a `Library` class with:
- Static variable `totalBooks` to track all books across all libraries
- Static method `getTotalBooks()` to return the total
- Instance variables: `libraryName`, `books` (array of Strings), `bookCount`
- Methods: `addBook(String bookTitle)`, `removeBook(String bookTitle)`, `displayBooks()`
- Use `this` keyword appropriately throughout

### Exercise 8: Complex Constructor Chaining and Method Parameters
**Question:** Create a `Car` class with:
- Private variables: `make`, `model`, `year`, `mileage`, `fuelLevel`
- Multiple constructors using constructor chaining
- Method `drive(int miles)` that increases mileage and decreases fuel
- Method `refuel()` and overloaded `refuel(double gallons)`
- Method `getCarInfo()` that returns detailed information
- Proper use of `this` keyword and access modifiers

### Exercise 9: Advanced Challenge - Complete Application
**Question:** Create a complete `StudentManagement` system with:
- `Student` class with private fields and proper encapsulation
- Static variable to track total students created
- `StudentManager` class to handle multiple students
- Demonstrate all concepts learned: constructors, methods, static members, access control, object creation and manipulation

## Instructions:
1. Work through the exercises in order from Easy to Hard
2. Each exercise builds upon the concepts from previous ones
3. Focus on applying the concepts you've learned this week
4. Test your solutions by creating objects and calling methods
5. Pay attention to proper use of access modifiers, constructors, and the `this` keyword
